```python
import pandas as pd

file1 = pd.read_csv('./file1.csv')
file2 = pd.read_csv('./file2.csv')
file3 = pd.read_csv('./file3.csv')
```

```python
display(file1.shape)
display(file2.shape)
display(file3.shape)
```

```python
# shows all column values
display(file1.columns)
display(file2.columns)
display(file3.columns)
```

```python
# converting every column in lower case
cols_1 = []
for column in file1.columns:
    cols_1.append(column.lower())
file1.columns = cols_1

cols_2 = []
for column in file2.columns:
    cols_2.append(column.lower())
file2.columns = cols_2

cols_3 = []
for column in file3.columns:
    cols_3.append(column.lower())
file3.columns = cols_3

display(file1.columns)
display(file2.columns)
display(file3.columns)
```

```python
# renaming columns

file3 = file3.rename(columns={'state':'st'})
```

```python
# rearranging columns

file3 = file3[['customer', 'st', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'total claim amount', 'policy type', 'vehicle class']]

display(file1.columns)
display(file2.columns)
display(file3.columns)
```

```python
# concatenate the 3 dataframes

beeeg_file = pd.concat([file1, file2,file3], axis=0)
beeeg_file
```

```python
# data cleaning: columns education, open complaints

beeeg_file = beeeg_file.drop(['education', 'number of open complaints'], axis=1)
beeeg_file.head()
```

```python
# data cleaning: changing costumter lifetime value

beeeg_file['customer lifetime value'] = beeeg_file['customer lifetime value']*100
beeeg_file['customer lifetime value'] =  pd.to_numeric(beeeg_file['customer lifetime value'], errors='coerce')


#beeeg_file.dtypes
beeeg_file


```
