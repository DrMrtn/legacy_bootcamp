```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score
```

```python
X = feat_model.drop(['SURVEY_SCORE__C'],axis=1)
y = feat_model.SURVEY_SCORE__C
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
num_train = X_train.select_dtypes(np.number)
num_test = X_test.select_dtypes(np.number)
```

```python
transformer = MinMaxScaler().fit(num_train)
num_train_minmax = transformer.transform(num_train)
num_test_minmax = transformer.transform(num_test)
pd.DataFrame(num_train_minmax).head()
```

```python
# transforming cats
cat_train= X_train.select_dtypes(object)
cat_test= X_test.select_dtypes(object)

```

```python
pd.DataFrame(X_train)
```

```python
cat_train_ohe = cat_train.TOPIC__C
cat_test_ohe = cat_test.TOPIC__C
```

```python
cat_train_ohe = cat_train_ohe.array.reshape(-1, 1) 
cat_test_ohe = cat_test_ohe.array.reshape(-1, 1)
```

```python
ohe = OneHotEncoder(handle_unknown='error', drop='first').fit(cat_train_ohe) 
train_ohe = ohe.transform(cat_train_ohe).toarray()
test_ohe = ohe.transform(cat_test_ohe).toarray()
pd.DataFrame(test_ohe)
```

```python
cat_train_ord = cat_train.drop(['TOPIC__C'],axis=1)
cat_test_ord = cat_test.drop(['TOPIC__C'],axis=1)
```

```python
display(feat_model.TYPE_OF_CHANGE__C.unique())
display(feat_model.CURRENT_ANSWER__C.unique())
display(feat_model.ACCOUNT_CUSTOMER_TYPE__C.unique())
```

```python
cat_train
```

```python
cat_train_ord["TYPE_OF_CHANGE__C"] = cat_train_ord["TYPE_OF_CHANGE__C"].map({"No Change" : 0, "Unknown" : 0, "Downgrade" : 0, "New" : .5, "Upgrade" : 1})
cat_train_ord["CURRENT_ANSWER__C"] = cat_train_ord["CURRENT_ANSWER__C"].map({"Unknown" : 0, "This is not done" : 0, "We do not do this consistently" : .5,                                                                             "Other" : 1, "Slight improvement" : .5, "This is consistently done" : 1})
cat_train_ord["ACCOUNT_CUSTOMER_TYPE__C"] = cat_train_ord["ACCOUNT_CUSTOMER_TYPE__C"].map({"Prospect" : 0, "New Business" : .5, "Existing Business" : 1})


cat_test_ord["TYPE_OF_CHANGE__C"] = cat_test_ord["TYPE_OF_CHANGE__C"].map({"No Change" : 0, "Unknown" : 0, "Downgrade" : 0, "New" : .5, "Upgrade" : 1})
cat_test_ord["CURRENT_ANSWER__C"] = cat_test_ord["CURRENT_ANSWER__C"].map({"Unknown" : 0, "This is not done" : 0, "We do not do this consistently" : .5,                                                                             "Other" : 1, "Slight improvement" : .5, "This is consistently done" : 1})
cat_test_ord["ACCOUNT_CUSTOMER_TYPE__C"] = cat_test_ord["ACCOUNT_CUSTOMER_TYPE__C"].map({"Prospect" : 0, "New Business" : .5, "Existing Business" : 1})

```

```python
cat_train = np.concatenate((train_ohe,cat_train_ord),axis=1)
cat_test = np.concatenate((test_ohe,cat_test_ord),axis=1)
pd.DataFrame(cat_train)
```

```python
X_train = np.concatenate((num_train_minmax,cat_train),axis=1)
X_test = np.concatenate((num_test_minmax,cat_test),axis=1)
pd.DataFrame(X_test)
```

```python
model1 = DecisionTreeRegressor()
model2 = LinearRegression()
model3 = KNeighborsRegressor()


model_pipeline = [model1, model2, model3]
model_names = ['Decision Tree Regressor', 'Linear Regression', 'KNN']
scores = {}
for model, model_name in zip(model_pipeline, model_names):
    mean_score = np.mean(cross_val_score(model, X_train, y_train, cv=5))
    scores[model_name] = mean_score
print(scores)
######
val_scores = {}
for model, model_name in zip(model_pipeline,model_names):
    model.fit(X_train, y_train)
    val_scores[model_name] = model.score(X_test,y_test)
print(val_scores)
```
