```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
nums = pd.read_csv('files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
nums = nums.select_dtypes(np.number)
nums.columns = [column.lower() for column in nums.columns]
nums.columns = nums.columns.str.replace(" ","_")
nums
```

```python
for col in nums.columns:
    sns.displot(data=nums,x=col)
    sns.set(rc = {'figure.figsize':(8,6)})
    plt.show()
```

```python
def log_transfom_clean1(x):
    x = np.log10(x)
    if np.isfinite(x):
        return x
    else:
        return 0
```

```python
nums.customer_lifetime_value = nums.customer_lifetime_value.apply(log_transfom_clean1)
sns.displot(data=nums,x=nums.customer_lifetime_value)
```

```python
nums.income = nums.income.apply(log_transfom_clean1)
sns.displot(data=nums,x=nums.income)
```

```python
nums.monthly_premium_auto = nums.monthly_premium_auto.apply(log_transfom_clean1)
sns.displot(data=nums,x=nums.monthly_premium_auto)
```

```python
nums.total_claim_amount = nums.total_claim_amount.apply(log_transfom_clean1)
sns.displot(data=nums,x=nums.total_claim_amount)
```
